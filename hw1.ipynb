将数组的前 2/3 部分排序<br>
将数组的后 2/3 部分排序<br>

通过三次递归调用，确保整个数组有序。<br>

````python
# add your code here
# 3.使用python编程实现上述算法或其对应的更高效版本的算法
import math

def stooge_sort(arr, p, r):
    n = r - p + 1
    # 长度为2且逆序则交换
    if n == 2 and arr[p] > arr[r]:
        arr[p], arr[r] = arr[r], arr[p]
        return
    # 长度 >= 3 时递归
    if n >= 3:
        # 取前 2/3 的长度向上取整
        m = math.ceil(2 * n / 3)
        stooge_sort(arr, p, p + m - 1)      # 前 m 个元素
        stooge_sort(arr, r - m + 1, r)      # 后 m 个元素
        stooge_sort(arr, p, p + m - 1)      # 再次处理前 m 个元素
def do_something(arr):
    stooge_sort(arr, 0, len(arr) - 1)


if __name__ == "__main__":
    test_arr = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
    print("原数组:", test_arr)
    do_something(test_arr)
    print("排序后:", test_arr)

# your algorithm time complexity is:
# O(n^2)
````

e.g. :

profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:<br>
注：本题参考大语言模型<br>
1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。<br>
本题参考大语言模型<br>
该算法采用滑动窗口 / 动态规划（Kadane 算法变体）策略：<br>
转化问题：输入的 profits 数组是每日收益的涨跌值。在第 $i$ 天买入并在第 $j$ 天卖出，总利润等于区间 profits[i : j] 的连续子数组之和。因此，寻找最大利润等价于寻找该数组的最大子数组和。<br>
状态更新与贪心选择：单遍遍历数组，用变量 current_sum 维护当前连续区间的利润累加和：<br>
如果 current_sum 变成负数，说明前面的区间对后续利润只有负贡献。此时立即舍弃前面区间，将当天作为新的买入起点重新开始累加。<br>
否则，继续将当天收益累加进当前区间。<br>
记录全局最优：在遍历过程中，每当 current_sum 超过历史最大值时，动态更新全局最大利润，并同步记录当前的买入日期（窗口起点）和卖出日期（当前遍历位置）。<br>



````python
# add your code here
# 1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。
def find_max_profit(profits):
    if not profits:
        return 0, 0, 0

    max_profit = float('-inf')
    current_sum = 0
    
    start = 0
    buy_day = 0
    sell_day = 0
    
    for i in range(len(profits)):
        if current_sum < 0:
            current_sum = profits[i]
            start = i
        else:
            current_sum += profits[i]
            
        if current_sum > max_profit:
            max_profit = current_sum
            buy_day = start + 1
            sell_day = i + 1
            
    return max_profit, buy_day, sell_day

# 测试样例
if __name__ == "__main__":
    profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
    max_prof, buy, sell = find_max_profit(profits)
    
    print(f"最大利润: {max_prof}")
    print(f"买入日期: 第 {buy} 天")
    print(f"卖出日期: 第 {sell} 天")

# your algorithm time complexity is:
# O(n)
````

1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现

idea:<br>
1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现<br>
该算法采用的是线性扫描策略。<br>
区间限制：根据题目对“局部最小”的定义，该元素必须比两边的邻居都小。因此，算法只遍历数组内部的元素，即下标从 1 到 n - 2。<br>
条件判断：在循环中，逐个检查当前元素 arr[i] 是否同时小于其左侧邻居 arr[i-1] 和右侧邻居 arr[i+1]。<br>
结果收集：若满足上述条件，则说明找到了一个局部最小值，将其加入结果列表 minima 中。遍历结束后返回所有找到的值。<br>

````python
# add your code here
# 1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
def find_all_local_minima(arr):
    #返回数组中所有局部最小元素的值
    n = len(arr)
    minima = []
    for i in range(1, n - 1):
        if arr[i] < arr[i-1] and arr[i] < arr[i+1]:
            minima.append(arr[i])
    return minima

if __name__ == "__main__":
    A = [9, 3, 7, 2, 1, 4, 5]
    result = find_all_local_minima(A)
    print(f"局部最小元素为: {result}")

# your algorithm time complexity is:
# O(n)
````

idea:<br>

2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现<br>
采用递增建堆与弹出策略。从空堆开始，通过连续 $n$ 次插入元素（向上调整）建立一个最小堆。建堆完成后，连续调用 $k$ 次弹顶操作（Extract-min / 向下调整），依次获取并输出当前堆中的最小值。<br>
<br>

3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。<br>
注：本题参考大语言模型<br>
采用数据流 Top-K / 固定容量堆策略。<br>
初始化：先读取数据流的前 $k$ 个元素，直接建立一个容量为 $k$ 的最大堆（堆顶为当前前 $k$ 个数里的最大值）。<br>
动态更新：逐个读取后续元素，若新元素小于堆顶，则用该新元素替换堆顶并向下调整；若大于等于堆顶则直接丢弃。<br>
输出：遍历完所有数据后，堆内留下的即为最小的 $k$ 个数，最后对这 $k$ 个数排序输出。该方法将内存占用降至 $O(k)$，完美解决外存海量数据限制。<br>


````python
# add your code here
import heapq

# 题目 1：堆排序实现并输出前 k 个最小值
def _heapify(arr, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2
    if l < n and arr[i] < arr[l]:
        largest = l
    if r < n and arr[largest] < arr[r]:
        largest = r
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        _heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)
    # 建最大堆
    for i in range(n // 2 - 1, -1, -1):
        _heapify(arr, n, i)
    # 一个个从堆顶取出元素
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        _heapify(arr, i, 0)

def solve_method_1(arr, k):
    """
    时间复杂度: O(n log n)
    """
    arr_copy = arr.copy()
    heap_sort(arr_copy)
    return arr_copy[:k]


# 题目 2：不断插入建最小堆，并调用 k 次 Extract-min
def solve_method_2(arr, k):
    """
    时间复杂度: O(n log n)
    """
    min_heap = []
    # 往空堆中不断插入元素建堆
    for num in arr:
        heapq.heappush(min_heap, num)
    
    # 调用 k 次 Extract-min (heappop)
    result = [heapq.heappop(min_heap) for _ in range(k)]
    return result


# 题目 3：海量数据/数据流场景，使用大小为 k 的最大堆
def solve_method_3(data_stream, k):
    """
    时间复杂度: O(n log k)
    空间复杂度: O(k)
    """
    # 模拟最大堆（通过将数值取负实现）
    max_heap = []
    
    for i, num in enumerate(data_stream):
        if i < k:
            heapq.heappush(max_heap, -num)
        else:
            # 如果当前数字小于堆顶的最大值，则替换堆顶
            if -num > max_heap[0]:
                heapq.heapreplace(max_heap, -num)
                
    # 恢复原数值并按从小到大排序输出
    result = [-x for x in max_heap]
    result.sort()
    return result


# 测试运行
if __name__ == "__main__":
    test_arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
    k_value = 3
    
    print("--- 原始数组 ---")
    print(test_arr)
    print(f"目标：找出前 {k_value} 个最小的数字\n")
    
    print("题目1结果")
    print(solve_method_1(test_arr, k_value))
    
    print("\n题目2结果")
    print(solve_method_2(test_arr, k_value))
    
    print("\n题目3结果")
    print(solve_method_3(test_arr, k_value))

# your algorithm time complexity is:
# 第一问：O(nlogn)
# 第二问：O(nlogn)
# 第三问：O(nlogk)
````

>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:<br>
注：本题参考大语言模型<br>
该问题属于典型的碰撞模拟问题，适合采用栈（Stack）数据结构来解决：<br>
碰撞条件：只有当栈顶星球向右移动（正数），且当前遍历的星球向左移动（负数）时，两颗星球才会发生碰撞。其他情况（如同向运动、或左负右正）均不会相撞，直接将当前星球压入栈中。<br>
碰撞模拟：当发生碰撞时，通过循环比较栈顶星球与当前星球的绝对值：<br>
当前星球质量大：栈顶星球消失（弹出栈），继续与新的栈顶星球比较。<br>
当前星球质量小：当前星球消失，停止比较，处理下一个星球。<br>
质量相同：根据题目规则（“自右向左的消失，自左向右的保持不变”），负数（自右向左）的星球消失，正数（自左向右）的栈顶星球保留。此时当前负数星球消失，停止比较。<br>
最终输出：遍历完所有星球后，栈中剩余的元素即为最终存续的星球状态。<br>


````python
# add your code here
def planet_collision(A):
    stack = []
    
    for planet in A:
        # 只有栈顶向右(>0)且当前向左(<0)时才会发生碰撞
        while stack and stack[-1] > 0 and planet < 0:
            top_mass = abs(stack[-1])
            curr_mass = abs(planet)
            
            if top_mass < curr_mass:
                # 栈顶星球质量小，被毁灭，继续与前一个星球碰撞
                stack.pop()
                continue
            elif top_mass == curr_mass:
                # 质量相同，自右向左的(当前负数)消失，自左向右的(栈顶正数)保留
                # 当前星球被毁灭，跳出循环
                break
            else:
                # 栈顶星球质量大，当前星球被毁灭，跳出循环
                break
        else:
            # 如果没有发生碰撞，或者碰撞中当前星球胜出，则入栈
            stack.append(planet)
            
    return stack

if __name__ == "__main__":
    example_1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
    example_2 = [23, -8, 9, -3, -7, 9, -23, 22]
    
    print(f"样例 1 输入: {example_1}")
    print(f"样例 1 输出: {planet_collision(example_1)}")
    print(f"样例 2 输入: {example_2}")
    print(f"样例 2 输出: {planet_collision(example_2)}")

# your algorithm time complexity is:
# O(n)
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:<br>
本题参考大语言模型<br>
采用动态规划策略来解决：<br>
1. 状态定义：设 $dp[j]$ 表示当前背包容量（小推车承重）为 $j$ 时，能装下大理石的最大价值。<br>
2. 状态转移方程：对于每块大理石（重量为 $w$，价值为 $v$），我们需要决定是“不选”还是“选择”它：<br>
$dp[j] = \max(dp[j], dp[j - w] + v)$<br>
3. 遍历顺序：为了确保每块大理石只能被选择一次（0-1 特性），在更新容量 $j$ 时，需要从最大承重 $W$ 从后往前（逆序）遍历到 $w$。<br>
4. 求解结果：最终 $dp[W]$ 即为承重为 $W$ 时的最大总价值。<br>

````python
# add your code here
def knapsack_01(stones, max_weight):
    # 初始化 dp 数组，容量为 0 到 max_weight
    dp = [0] * (max_weight + 1)
    # 遍历每一块大理石 (weight, value)
    for weight, value in stones:
        # 逆序遍历容量，确保每件物品只选择一次
        for j in range(max_weight, weight - 1, -1):
            dp[j] = max(dp[j], dp[j - weight] + value)
            
    return dp[max_weight]

# --- 测试验证 ---
if __name__ == "__main__":
    stones = [(5, 10), (4, 40), (6, 30), (3, 50)]
    max_weight = 9
    
    max_value = knapsack_01(stones, max_weight)
    print(f"小推车装大理石的最大价值为: {max_value}")
    
# your algorithm time complexity is:
# O(n*W),n为大理石的数量,W为小推车的最大承重
````

## 问题 8

给你二叉搜索树的根节点 root ，该树中的恰好两个节点的值被错误地交换。请在不改变其结构的情况下，恢复这棵树 。设计算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

e.g.:
---
1.

<div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入：root = [1,3,null,null,2]， 输出为：[3,1,null,null,2]， 

解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。

2.

 <div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

idea:<br>
1. 进行中序遍历，在遍历过程中记录前一个节点。
2. 如果当前节点值小于前一个节点值，说明找到了一个异常点：<br>
第一次异常：前一个节点就是第一个错位节点（first）。<br>
第二次异常：当前节点就是第二个错位节点（second）。
3. 遍历结束后交换 first 和 second 的值。

````python
# add your code here
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def recoverTree(self, root: TreeNode) -> None:
        # 恢复二叉搜索树，不改变结构，只交换两个错误节点的值
        self.first = None    # 第一个错误节点
        self.second = None   # 第二个错误节点
        self.prev = None     # 中序遍历的前一个节点
        
        def inorder(node):
            if not node:
                return
            
            # 遍历左子树
            inorder(node.left)
            
            # 检查当前节点
            if self.prev and self.prev.val > node.val:
                if not self.first:
                    self.first = self.prev  # 第一次异常，记录前一个节点
                self.second = node          # 第二次异常（或相邻情况）记录当前节点
            
            self.prev = node
            
            # 遍历右子树
            inorder(node.right)
        
        # 执行中序遍历
        inorder(root)
        
        # 交换两个错误节点的值
        if self.first and self.second:
            self.first.val, self.second.val = self.second.val, self.first.val

def list_to_tree(arr, index=0):
    # 将列表转换为二叉树（层序）
    if index >= len(arr) or arr[index] is None:
        return None
    root = TreeNode(arr[index])
    root.left = list_to_tree(arr, 2 * index + 1)
    root.right = list_to_tree(arr, 2 * index + 2)
    return root

def tree_to_list(root):
    # 将二叉树转换为列表表示（层序
    if not root:
        return []
    
    result = []
    queue = [root]
    
    while queue:
        node = queue.pop(0)
        if node:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)
    
    # 移除末尾的None
    while result and result[-1] is None:
        result.pop()
    
    return result

if __name__ == "__main__":
    solution = Solution()
    
    # 测试用例1
    print("测试用例1:")
    print(f"输入：root = [1,3,null,null,2]")
    tree1 = list_to_tree([1,3,None,None,2])
    solution.recoverTree(tree1)
    print(f"输出：{tree_to_list(tree1)}")
    print()
    
    # 测试用例2
    print("测试用例2:")
    print(f"输入：root=[3,1,4,null,null,2]")
    tree2 = list_to_tree([3,1,4,None,None,2])
    solution.recoverTree(tree2)
    print(f"输出：{tree_to_list(tree2)}")

# your algorithm time complexity is:
# O(n)
````

## 问题 10

给你一个二叉树的根节点 root ，树中每个节点都存放有一个 0 到 9 之间的数字。每条从根节点到叶节点的路径都代表一个数字：

> 例如，从根节点到叶节点的路径 1 -> 2 -> 3 表示数字 123 。

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

e.g.:
---
1.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-1.jpg"> </div>

 输入：root = [1,2,3], 输出：25

解释：从根到叶子节点路径 1->2 代表数字 12，1->3 代表数字 13，因此和为 12 + 13 = 25 。

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

idea:<br>
用 DFS从根遍历到叶子。<br>
维护当前路径形成的数字 current_sum。<br>
每走到一个节点，更新数字：current_sum = current_sum * 10 + node.val。<br>
遇到叶子节点时，把 current_sum 累加到总和中。<br>
回溯。<br>
最终返回总和<br>

````python
# add your code here
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def sumNumbers(self, root: TreeNode) -> int:
        def dfs(node, current_sum):
            if not node:
                return 0
            current_sum = current_sum * 10 + node.val
            # 如果是叶子节点
            if not node.left and not node.right:
                return current_sum
            # 否则继续递归
            return dfs(node.left, current_sum) + dfs(node.right, current_sum)
        
        return dfs(root, 0)

def list_to_tree(arr, index=0):
    if index >= len(arr) or arr[index] is None:
        return None
    root = TreeNode(arr[index])
    root.left = list_to_tree(arr, 2 * index + 1)
    root.right = list_to_tree(arr, 2 * index + 2)
    return root

if __name__ == "__main__":
    solution = Solution()
    
    # 示例 1
    root1 = list_to_tree([1, 2, 3])
    print(f"输入：root = [1,2,3]")
    print(f"输出：{solution.sumNumbers(root1)}")
    print()
    
    # 示例 2
    root2 = list_to_tree([4, 9, 0, 5, 1])
    print(f"输入：root = [4,9,0,5,1]")
    print(f"输出：{solution.sumNumbers(root2)}")

# your algorithm time complexity is:
# O(n)
````
