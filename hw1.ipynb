## 问题 1

根据下列递归公式, 计算下列 $T(n)$ 对应的的渐近上界。要求所求的边界尽可能的紧（tight）, 请写明步骤。

1. $T(1)=1; T(n)=T(n/4)+1$ for $n>1$
2. $T(1)=1;T(n)=3T(n/3)+n^2$ for $n>1$
3. $T(1)=1;T(n)=T(2n/3)+1$ for $n>1$
4. $T(1)=1;T(n)=5T(n/4)+n$ for $n>1$
5. $T(n)=1 \ for\ n \le 2 ; T(n)=T(\sqrt{n})+1 \ for \ n>2$

idea:

````python
# add your code here
import math
from functools import lru_cache

# 递归式 1: T(n) = T(n//4) + 1
@lru_cache(maxsize=None)
def T1(n):
    if n <= 1:
        return 1
    return T1(n // 4) + 1

# 递归式 2: T(n) = 3*T(n//3) + n**2
@lru_cache(maxsize=None)
def T2(n):
    if n <= 1:
        return 1
    return 3 * T2(n // 3) + n * n

# 递归式 3: T(n) = T(2*n//3) + 1
@lru_cache(maxsize=None)
def T3(n):
    if n <= 1:
        return 1
    return T3((2 * n) // 3) + 1

# 递归式 4: T(n) = 5*T(n//4) + n
@lru_cache(maxsize=None)
def T4(n):
    if n <= 1:
        return 1
    return 5 * T4(n // 4) + n

# 递归式 5: T(n) = T(isqrt(n)) + 1  (n>2)
@lru_cache(maxsize=None)
def T5(n):
    if n <= 2:
        return 1
    return T5(math.isqrt(n)) + 1

# 测试并打印结果
if __name__ == "__main__":
    print("渐近上界分析结果：")
    print("1. T(n) = T(n/4)+1  ->  Θ(log n)")
    print("2. T(n) = 3T(n/3)+n^2 -> Θ(n^2)")
    print("3. T(n) = T(2n/3)+1 -> Θ(log n)")
    print("4. T(n) = 5T(n/4)+n -> Θ(n^{log_4 5}) ≈ Θ(n^{1.161})")
    print("5. T(n) = T(√n)+1 -> Θ(log log n)\n")

    # 选取合适的 n 值进行数值验证
    ns1 = [4**i for i in range(1, 8)]          # 4 的幂
    ns2 = [3**i for i in range(1, 6)]          # 3 的幂
    ns3 = [int((3/2)**i) for i in range(1, 15)] # (3/2)^i 取整
    ns4 = [4**i for i in range(1, 8)]          # 4 的幂
    ns5 = [2**(2**i) for i in range(1, 6)]     # 2^(2^i)

    print("=== 递归1: T(n)=T(n/4)+1 ===")
    for n in ns1:
        print(f"n={n:8d}, T={T1(n):4d}, log4(n)={math.log(n,4):.2f}")

    print("\n=== 递归2: T(n)=3T(n/3)+n^2 ===")
    for n in ns2:
        print(f"n={n:8d}, T={T2(n):12d}, n^2={n*n:12d}")

    print("\n=== 递归3: T(n)=T(2n/3)+1 ===")
    for n in ns3:
        print(f"n={n:8d}, T={T3(n):4d}, log3/2(n)={math.log(n,1.5):.2f}")

    print("\n=== 递归4: T(n)=5T(n/4)+n ===")
    for n in ns4:
        ratio = T4(n) / (n ** (math.log(5,4)))
        print(f"n={n:8d}, T={T4(n):12d}, n^{{log4(5)}}={n**math.log(5,4):.2f}, ratio={ratio:.4f}")

    print("\n=== 递归5: T(n)=T(√n)+1 ===")
    for n in ns5:
        print(f"n={n:15d}, T={T5(n):4d}, log2(log2(n))={math.log2(math.log2(n)):.4f}")

# your algorithm time complexity is:
# O()
````

````python
# add your code here
import math

def stooge_sort(arr, p, r):
    n = r - p + 1
    # 基本情况：长度为2且逆序则交换
    if n == 2 and arr[p] > arr[r]:
        arr[p], arr[r] = arr[r], arr[p]
        return
    # 长度 >= 3 时递归
    if n >= 3:
        # 取前 2/3 的长度（向上取整）
        m = math.ceil(2 * n / 3)
        stooge_sort(arr, p, p + m - 1)      # 前 m 个元素
        stooge_sort(arr, r - m + 1, r)      # 后 m 个元素
        stooge_sort(arr, p, p + m - 1)      # 再次处理前 m 个元素
# 对外包装函数
def do_something(arr):
    stooge_sort(arr, 0, len(arr) - 1)


# 测试示例
if __name__ == "__main__":
    test_arr = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
    print("原数组:", test_arr)
    do_something(test_arr)
    print("排序后:", test_arr)


# your algorithm time complexity is:
#O(n2)
````

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:

````python
# add your code here
def max_profit(profits):
    n = len(profits)
    if n < 2:
        return 0, -1, -1  # 无法完成交易

    # 前缀和初始值
    min_prefix = 0        # 最小前缀和
    min_day = 0           # 最小前缀和对应的天数（前缀和索引）
    max_gain = -float('inf')
    buy_day = sell_day = -1

    prefix_sum = 0
    for j in range(1, n+1):
        prefix_sum += profits[j-1]   # 前j天的收益和
        gain = prefix_sum - min_prefix
        if gain > max_gain:
            max_gain = gain
            buy_day = min_day + 1
            sell_day = j
        if prefix_sum < min_prefix:
            min_prefix = prefix_sum
            min_day = j
    return max_gain, buy_day, sell_day

if __name__ == "__main__":
    profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
    gain, buy, sell = max_profit(profits)
    print(f"最大利润: {gain}")
    print(f"买入日期: 第{buy}天")
    print(f"卖出日期: 第{sell}天")


# your algorithm time complexity is:
# O(1)
````

1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现

````python
# add your code here
def find_all_local_minima(arr):
    """返回数组中所有局部最小元素的值（按出现顺序）"""
    n = len(arr)
    minima = []
    for i in range(1, n - 1):
        if arr[i] < arr[i-1] and arr[i] < arr[i+1]:
            minima.append(arr[i])
    return minima


if __name__ == "__main__":
    A = [9, 3, 7, 2, 1, 4, 5]
    result = find_all_local_minima(A)
    print(f"局部最小元素为: {result}")
    
# your algorithm time complexity is:
# O(n)
````

## 问题 5

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

1. 将n个数先进行排序, 然后按顺序输出最小的k个数。要求：选择合适的排序算法实现上述操作, 计算算法时间复杂度, 并使用python编程实现。
2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现
3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

idea:

````python
# add your code here
import heapq

# 第一部分：先排序后输出最小的 k 个数
def part1_sort(arr, k):
    """
    先排序后输出最小的 k 个数
    时间复杂度: O(n log n)
    """
    if k <= 0:
        return []
    sorted_arr = sorted(arr)
    return sorted_arr[:k]


# 第二部分：建堆 + k 次 Extract-min 
class MinHeap:
    """最小堆（手动实现）"""
    def __init__(self):
        self.heap = []

    def parent(self, i):
        return (i - 1) // 2

    def left_child(self, i):
        return 2 * i + 1

    def right_child(self, i):
        return 2 * i + 2

    def swap(self, i, j):
        self.heap[i], self.heap[j] = self.heap[j], self.heap[i]

    def heapify_up(self, i):
        """向上调整"""
        while i > 0 and self.heap[i] < self.heap[self.parent(i)]:
            self.swap(i, self.parent(i))
            i = self.parent(i)

    def heapify_down(self, i):
        """向下调整"""
        n = len(self.heap)
        while True:
            smallest = i
            left = self.left_child(i)
            right = self.right_child(i)
            if left < n and self.heap[left] < self.heap[smallest]:
                smallest = left
            if right < n and self.heap[right] < self.heap[smallest]:
                smallest = right
            if smallest == i:
                break
            self.swap(i, smallest)
            i = smallest

    def insert(self, val):
        """插入元素"""
        self.heap.append(val)
        self.heapify_up(len(self.heap) - 1)

    def extract_min(self):
        """弹出最小值"""
        if not self.heap:
            raise IndexError("Heap is empty")
        min_val = self.heap[0]
        last = self.heap.pop()
        if self.heap:
            self.heap[0] = last
            self.heapify_down(0)
        return min_val

    def size(self):
        return len(self.heap)


def part2_heap_insert(arr, k):
    if k <= 0:
        return []

    heap = MinHeap()
    # 建堆：逐个插入 O(n log n)
    for val in arr:
        heap.insert(val)

    # 提取 k 次 O(k log n)
    result = []
    for _ in range(min(k, heap.size())):
        result.append(heap.extract_min())
    return result


# 第三部分：海量数据（只能存储 k 个数据）
def part3_large_data(data_stream, k):
    if k <= 0:
        return []

    max_heap = []
    for val in data_stream:
        if len(max_heap) < k:
            heapq.heappush(max_heap, -val)
        else:
            # 如果当前值比堆顶（当前最大的候选值）小，则替换
            if val < -max_heap[0]:
                heapq.heapreplace(max_heap, -val)

    # 将堆中的负数转回正数，并升序输出
    result = sorted([-x for x in max_heap])
    return result


#测试代码
if __name__ == "__main__":
    arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
    k = 3

    print("原数组:", arr)
    print("需要最小的", k, "个数")

    # 第一部分测试
    res1 = part1_sort(arr, k)
    print("\n【第一部分：排序法】")
    print("结果:", res1)

    # 第二部分测试
    res2 = part2_heap_insert(arr, k)
    print("\n【第二部分：插入建堆 + k次Extract-min】")
    print("结果:", res2)

    # 第三部分测试：将 arr 视为数据流（模拟海量数据）
    def data_stream():
        for v in arr:
            yield v

    res3 = part3_large_data(data_stream(), k)
    print("\n【第三部分：海量数据算法（只能存储 k 个数据）】")
    print("结果:", res3)


# your algorithm time complexity is:
# O(Nlogk)
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:

````python
# add your code here
def asteroid_collision(asteroids):
    stack = []
    for cur in asteroids:
        while stack and stack[-1] > 0 and cur < 0:
            top = stack[-1]
            # 比较质量
            if abs(top) > abs(cur):
                # 当前星球消失，不加入栈
                break
            elif abs(top) < abs(cur):
                # 栈顶星球消失，继续与新的栈顶比较
                stack.pop()
            else:
                # 质量相等，向左运动的星球消失（即当前星球消失），栈顶保留
                stack.pop()
                break
        else:
            # 没有发生碰撞或碰撞结束后当前星球仍然存活，则入栈
            stack.append(cur)
    return stack


if __name__ == "__main__":
    # 示例 1
    A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
    result1 = asteroid_collision(A1)
    print("输入:", A1)
    print("输出:", result1)  # 期望 [-3, -6, 2, 8, 9, 1]

    # 示例 2
    A2 = [23, -8, 9, -3, -7, 9, -23, 22]
    result2 = asteroid_collision(A2)
    print("\n输入:", A2)
    print("输出:", result2)  # 期望 [23, 22]

# your algorithm time complexity is:
# O(n)
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

````python
# add your code here
def knapsack_01(weights, values, capacity):
    """
    0-1背包问题动态规划解法
    :param weights: 物品重量列表
    :param values:  物品价值列表
    :param capacity: 背包最大承重
    :return: 最大价值
    """
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        w = weights[i - 1]
        v = values[i - 1]
        for c in range(capacity + 1):
            if c < w:
                # 当前物品放不下，只能不选
                dp[i][c] = dp[i - 1][c]
            else:
                # 不选 vs 选
                dp[i][c] = max(dp[i - 1][c], dp[i - 1][c - w] + v)

    return dp[n][capacity]


if __name__ == "__main__":
    # 输入数据
    items = [(5, 10), (4, 40), (6, 30), (3, 50)]
    weights = [w for w, _ in items]
    values = [v for _, v in items]
    capacity = 9

    max_value = knapsack_01(weights, values, capacity)
    print("最大价值为:", max_value)


# your algorithm time complexity is:
# O(n)
````

## 问题 8

给你二叉搜索树的根节点 root ，该树中的恰好两个节点的值被错误地交换。请在不改变其结构的情况下，恢复这棵树 。设计算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

e.g.:
---
1.

<div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入：root = [1,3,null,null,2]， 输出为：[3,1,null,null,2]， 

解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。

2.

 <div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

idea:

````python
# add your code here
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def recover_tree(root):
    # 用于中序遍历的栈
    stack = []
    # 记录前一个访问的节点
    prev = None
    # 记录两个错误的节点
    first = second = None

    # 迭代中序遍历
    node = root
    while node or stack:
        while node:
            stack.append(node)
            node = node.left
        node = stack.pop()

        # 检查是否出现逆序
        if prev and prev.val > node.val:
            if first is None:
                first = prev
            second = node
        prev = node

        node = node.right

    # 交换两个错误节点的值
    if first and second:
        first.val, second.val = second.val, first.val


# 构建树
def build_tree_from_list(lst, index=0):
    """根据列表构建二叉树（层序），None 表示空节点。"""
    if index >= len(lst) or lst[index] is None:
        return None
    root = TreeNode(lst[index])
    root.left = build_tree_from_list(lst, 2 * index + 1)
    root.right = build_tree_from_list(lst, 2 * index + 2)
    return root


def tree_to_list(root):
    """将二叉树转换为层序列表（包含 None），方便观察结果。"""
    if not root:
        return []
    result = []
    queue = [root]
    while queue:
        node = queue.pop(0)
        if node:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)
    # 去除末尾连续的 None
    while result and result[-1] is None:
        result.pop()
    return result


if __name__ == "__main__":
    # 示例1: root = [1,3,null,null,2]
    root1 = build_tree_from_list([1, 3, None, None, 2])
    print("恢复前 (层序):", tree_to_list(root1))
    recover_tree(root1)
    print("恢复后 (层序):", tree_to_list(root1))  # 期望 [3,1,null,null,2]

    print("\n" + "-" * 30 + "\n")

    # 示例2: root = [3,1,4,null,null,2]
    root2 = build_tree_from_list([3, 1, 4, None, None, 2])
    print("恢复前 (层序):", tree_to_list(root2))
    recover_tree(root2)
    print("恢复后 (层序):", tree_to_list(root2))  # 期望 [2,1,4,null,null,3]


# your algorithm time complexity is:
# O(n)
````

## 问题 10

给你一个二叉树的根节点 root ，树中每个节点都存放有一个 0 到 9 之间的数字。每条从根节点到叶节点的路径都代表一个数字：

> 例如，从根节点到叶节点的路径 1 -> 2 -> 3 表示数字 123 。

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

e.g.:
---
1.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-1.jpg"> </div>

 输入：root = [1,2,3], 输出：25

解释：从根到叶子节点路径 1->2 代表数字 12，1->3 代表数字 13，因此和为 12 + 13 = 25 。

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

idea:

````python
# add your code here
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def sum_numbers(root):
    def dfs(node, current_sum):
        if not node:
            return 0
        current_sum = current_sum * 10 + node.val
        if not node.left and not node.right:
            return current_sum
        return dfs(node.left, current_sum) + dfs(node.right, current_sum)

    return dfs(root, 0)


# 构建树
def build_tree_from_list(lst, index=0):
    """根据列表构建二叉树（层序），None 表示空节点。"""
    if index >= len(lst) or lst[index] is None:
        return None
    root = TreeNode(lst[index])
    root.left = build_tree_from_list(lst, 2 * index + 1)
    root.right = build_tree_from_list(lst, 2 * index + 2)
    return root


#测试
if __name__ == "__main__":
    root1 = build_tree_from_list([1, 2, 3])
    print("示例1 路径和:", sum_numbers(root1))  # 输出: 25

    root2 = build_tree_from_list([4, 9, 0, 5, 1])
    print("示例2 路径和:", sum_numbers(root2))  # 输出: 1026

# your algorithm time complexity is:
# O(n)
````
